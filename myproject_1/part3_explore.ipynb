# 用户词典调整分词结果

```python
jieba.load_userdict("air_property_value.txt")
```

```python
import re
corpus_clean = []
fil = re.compile(r'[^0-9.\u4e00-\u9fa5]+')
for raw in data_sale_24_month.loc[data_sale_24_month["三级品类名称"]=='空调']["商品名称"].values:
    raw = fil.sub('', str(raw))
    raw = raw.strip()
    corpus_clean.append(raw)
```

```python
corpus_clean
```

# 统计词频

```python
def fenci(data):
    cut_words = map(lambda s: list(jieba.cut(s)), data)
    return list(cut_words) 
# 以下为传入格式
#fenci(['美的2匹智行立柜式变频冷暖客厅圆柱空调柜机线下同款51284003',"'海尔空调7223显示样机演示机壳非卖品请勿拍'"])
```

```python
fenci(['美的2匹智行立柜式变频冷暖客厅圆柱空调柜机线下同款51284003',"'海尔空调7223显示样机演示机壳非卖品请勿拍'"])
```

```python
fenci_list = fenci(corpus_clean)
```

```python
from collections import defaultdict
frequency = defaultdict(int)

for text in fenci_list:
    for token in text:
        frequency[token] += 1
print(len(frequency))
frequency_sort = sorted(frequency.items(), key=lambda x: x[1], reverse=True)
frequency_sort = pd.DataFrame(frequency_sort, columns=["词语", "词频"])
print("打印词频:")
frequency_sort
```
