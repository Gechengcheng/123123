# 计算近一年和上一年二级品类销售额同比

```python
last_year_second_cate = data_cate_sale_24_month.loc[data_cate_sale_24_month["月份"]=="上1年"].groupby(['二级品类名称'])["销售额"].agg([np.sum]).rename(columns={"sum":"上一年"})
recent_year_second_cate = data_cate_sale_24_month.loc[data_cate_sale_24_month["月份"]=="近1年"].groupby(['二级品类名称'])["销售额"].agg([np.sum]).rename(columns={"sum":"近一年"})
```

```python
second_cate = pd.concat([last_year_second_cate,recent_year_second_cate],axis=1)
second_cate["同比"] = (second_cate["近一年"]-second_cate["上一年"])/second_cate["上一年"]
```

```python
second_cate["percent"] = second_cate["同比"].apply(lambda x: '%.2f%%' % (x*100))
```
